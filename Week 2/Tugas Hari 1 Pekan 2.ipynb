<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Jawab: 

Deskriptif statistik adalah sebuah metode untuk memberikan informasi dengan cara mengumpulkan data lalu menyajikan nya dalam suatu himpunan data [1]. Dengan adanya deskriptif statistik data-data tersebut akan tersaji ringkas dan rapih. deskriptif statistik terdiri dari pemusatan data, penyebaran data, serta kecendrungan dari suatu gugus data [2].

[1] Ronald E.Walpole. Pengantar Statistika, halaman 2-5". 1993. Jakarta: PT Gramedia Pustaka Utama.

[2] Dergibson Siagian & Sugiarto. Metode Statistika untuk Bisnis dan Ekonomi, halaman 4-6". 2002. Jakarta: PT Gramedia Pustaka Utama.

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
print('nilai rata rata kolom price:', diamonds_1['price'].mean())
print('nilai standar deviasi:', diamonds_1['depth'].std())
print('nilai maksimum dari carat:', diamonds_1['carat'].max())
```

* nilai standar deviasi: 1.4326213188337733
