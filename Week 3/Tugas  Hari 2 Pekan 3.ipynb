Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
#code here

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, cos, label='Cosx', color='red')
ax.plot(x, cos3, label='3 Cos(x)', color='green', linestyle = 'dashed')
ax.plot(x, sin, label='Sin(x-90)', color='blue', linestyle = 'dashdot')
ax.plot(x, sin2, label='2Sinx', color='#d0d046',linestyle = 'None', marker = 'x')


plt.title('Grafik Cos Sin')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.legend()
plt.show
```
